```python
import cv2 as cv
import os
import numpy as np
from matplotlib import pyplot as plt
path = "/media/rishabhstein/New Volume/CT_Images/Wierzbica/"
l = os.listdir(path)
```

```python
sl = [l[200], l[-1]]
d = [];
for i in sl:
    fname = path + i
    tmp = cv.imread(fname, cv.IMREAD_GRAYSCALE)
    d.append(tmp)

dims = np.shape(tmp)

img = np.reshape(d,[2,dims[0],dims[1]])
diff = img[1,:,:] - img[0,:,:]
plt.imshow(diff)
print(np.mean(np.abs(diff)))
```
